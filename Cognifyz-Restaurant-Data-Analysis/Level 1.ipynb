```python
cuisines = df['Cuisines'].str.split(', ')
cuisines.value_counts()
```

```python
all_cuisines = cuisines.explode()
```

```python
cuisine_counts = all_cuisines.value_counts()
print("Cuisine Counts :" ,cuisine_counts.sum())
```

```python
top_3 = cuisine_counts.head(3)

print("Top 3 Most Common Cuisines:")
print(top_3)
```

```python
total_restaurants = len(df)

percentage = (top_3 / total_restaurants) * 100

print("\nPercentage of Restaurants Serving Top Cuisines:")
print(percentage)
```

```python
plt.figure(figsize=(8,5))

top_3.plot(kind='bar')

plt.title("Top 3 Most Common Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")

plt.xticks(rotation=0)

plt.show()
```

# Level 1 → Task 2:City Analysis .

```python
df['City'].head(100)
```

```python
df['City'].value_counts()
```

```python
df.columns
```

```python
df = df.dropna(subset=['City', 'Aggregate rating'])
```

```python

city_counts = df['City'].value_counts()
```

```python
city_counts
```

```python
top_city = city_counts.idxmax()

top_city_count = city_counts.max()

print("City with Highest Number of Restaurants:")
print(top_city)

print("\nNumber of Restaurants:")
print(top_city_count)
```

```python
city_avg_rating = df.groupby('City')['Aggregate rating'].mean()
city_avg_rating
```

```python
city_avg_rating = city_avg_rating.sort_values(ascending=False)

print(city_avg_rating)
```

```python
highest_rated_city = city_avg_rating.idxmax()

highest_rating = city_avg_rating.max()

print("\nCity with Highest Average Rating:")
print(highest_rated_city)

print("\nHighest Average Rating:")
print(highest_rating)
```

```python
plt.figure(figsize=(10,6))

city_counts.head(10).plot(kind='bar')

plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,6))

city_avg_rating.head(10).plot(kind='bar')

plt.title("Top 10 Cities by Average Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")

plt.xticks(rotation=45)

plt.show()
```

# Level 1 → Task 3: Price Range Distribution

```python
df= pd.read_excel("C:/Users/Surya/Downloads/Dataset .xlsx")
```

```python
print(df.head())
print(df.columns)
```

```python
df = df.dropna(subset=['Price range'])
```

```python
price_counts = df['Price range'].value_counts().sort_index()

print(price_counts)
```

```python
total_restaurants = len(df)

price_percentage = (price_counts / total_restaurants) * 100

print("\nPercentage Distribution:")
print(price_percentage)
```

```python
plt.figure(figsize=(8,5))

price_counts.plot(kind='bar')

plt.title("Distribution of Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

plt.xticks(rotation=0)

plt.show()
```

# Level 1 → Task 4: Online Delivery Analysis

```python
df= pd.read_excel("C:/Users/Surya/Downloads/Dataset .xlsx")
```

```python
print(df.head())
print(df.columns)
```

```python
df = df.dropna(subset=['Has Online delivery', 'Aggregate rating'])
```

```python
delivery_counts = df['Has Online delivery'].value_counts()

print(delivery_counts)
```

```python
delivery_percentage = (delivery_counts / len(df)) * 100

print("\nPercentage of Restaurants Offering Online Delivery:")
print(delivery_percentage)
```

```python
avg_ratings = df.groupby('Has Online delivery')['Aggregate rating'].mean()

print("\nAverage Ratings:")
print(avg_ratings)
```

```python
plt.figure(figsize=(6,6))

delivery_counts.plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Online Delivery Availability")
plt.ylabel("")

plt.show()
```

```python
plt.figure(figsize=(6,5))

avg_ratings.plot(kind='bar')

plt.title("Average Ratings: Online Delivery vs No Delivery")
plt.xlabel("Online Delivery")
plt.ylabel("Average Rating")

plt.xticks(rotation=0)

plt.show()
```
